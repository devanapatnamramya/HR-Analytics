```python
#Importing DecisionTreeRegressor from sklearn.tree
from sklearn.tree import DecisionTreeRegressor
#Importing mean_squared_error,r2 score,mean_absolute_error from sklearn.metrics
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.metrics import r2_score
#Taking DecisiontreeRegressor to DTR with maximum depth =3
dtr = DecisionTreeRegressor(max_depth =3)
#fitting the x_train,y_train to the model
dtr.fit(x_train,y_train)
#predicting the results using x_train
y_pred = dtr.predict(x_train)
print(' r2 score for Training :',r2_score(y_train,y_pred)) #print of r2 score for train
print('Mean Absolute Error for Training: ',mean_absolute_error(y_train,y_pred))#Print MSE value for train
print('Mean Squared Error for Training: ',mean_squared_error(y_train,y_pred))  #Print MSE value for train
print('Root Mean Squared Error for Training: ',np.sqrt(mean_squared_error(y_train,y_pred)))  #Print RMSE value for train
print('---------------------------------------------')
#predicting the results using x_test
y_test_pred = dtr.predict(x_test)
print(' r2 score for Testing :',r2_score(y_test,y_test_pred)) #print of r2 score for test
print('Mean Absolute Error for Testing: ',mean_absolute_error(y_test,y_test_pred))#Print MSE value for test
print('Mean Squared Error for Testing: ',mean_squared_error(y_test,y_test_pred))  #Print MSE value for test
print('Root Mean Squared Error for Testing: ',np.sqrt(mean_squared_error(y_test,y_test_pred)))  #Print RMSE value for test
```

In this technique,we create random splits of data in the training-set manner and

## **STEP 8:HANDLING OUTLIERS**

**To handle the outliers here i am using IQR METHOD**

In this method by using **Inter Quartile Range(IQR)**, we detect outliers. IQR tells us the variation in the data set. Any value, which is beyond the range of -1.5 x IQR to 1.5 x IQR treated as outliers<br>
- Q1 represents the 1st quartile/25th percentile of the data.<br>
* Q2 represents the 2nd quartile/median/50th percentile of the data.<br>
- Q3 represents the 3rd quartile/75th percentile of the data.<br>
* (Q1–1.5IQR) represent the smallest value in the data set and (Q3+1.5IQR) represent the largest value in the data set
